## Fixes

### prefrontal_cortex metadata missing the content_type and size fields in ingest

Load all files for the original submission

```python
import requests as rq
files_sub = rq.get('https://api.ingest.archive.data.humancellatlas.org/submissionEnvelopes/5ee8d88f5e315a423810f7b3/files?size=80').json()
```

Get the path for all file descriptors:

```python
import os
descriptor_filenames = [] 
for root, dir, files in os.walk('descriptors'):
    descriptor_filenames.extend([os.path.join(root, file) for file in files])
```

Retrieve the descriptors for this specific dataset:

```python
descriptors_size_type = [descriptor for descriptor in descriptor_filenames if any([file['uuid']['uuid'] in descriptor for file in files_sub['_embedded']['files']])]
len(descriptors_size_type)
```

Retrieve the files from the newly created submission with file sizes:

```python
new_files = rq.get('https://api.ingest.staging.archive.data.humancellatlas.org/submissionEnvelopes/5f69d165b4b6b76768734377/files?size=80').json()
```

Create a map filename → size and schema_type:

```python
filename_to_size_type = {}
for file in new_files['_embedded']['files']:
    filename_to_size_type[file['fileName']] = {"size": file['size'], "fileContentType": file['fileContentType']}
```

Modify the 78 file descriptors

```python
import json
for descriptor_filename in descriptors_size_type:
    with open(descriptor_filename, 'r') as f:
        descriptor = json.load(f)
    print(f"Modifying size and content type for file {descriptor['file_name']}")
    from_map = filename_to_size_type[descriptor['file_name'].split('Z_')[-1]]
    descriptor['content_type'] = from_map['fileContentType']
    descriptor['size'] = from_map['size']
    print(f"\tNew content_type: {descriptor['content_type']}\n\tNew size: {descriptor['size']}\n\n")
    with open(descriptor_filename, 'w') as f:
        json.dump(descriptor, f)
```

### Missing field schema_type

Again, with the list of file descriptors:

```python
for descriptor in descriptor_filenames:
    with open(descriptor, 'r') as f:
        content = json.load(f)
    content['schema_type'] = 'file_descriptor'
    with open(descriptor, 'w') as f:
        json.dump(content, f)
```

### Version and crc32c

#### crc32c

With the previously created list of file_descriptors:

```python
for descriptor in descriptor_filenames:
    with open(descriptor, 'r') as f:
        content = json.load(f)
    content['crc32c'] = content['crc32c'].lower()
    with open(descriptor, 'w') as f:
        json.dump(content, f)
```


#### Version

We are going to change the every filename adding a timestamp with 6 decimals for the fraction of a second. Metadata within the files (except for the file descriptor) will remain unchanged.

The plan is:
1. Modify file_descriptors metadata to include 6d in `version` field
1. Modify file_descriptors metadata to point to file with 6d
1. Push all the changes to the staging bucket
1. Get a list of all filenames in staging bucket
1. Get a script to modify all filenames (gsutil mv <file with 3 digits> <file with 6 digits>)
1. Test first on local by copying the script without “gsutil”

Modify the file_descriptors metadata to include 6d

```python
for descriptor in descriptor_filenames:
    with open(descriptor, 'r') as f:
        content = json.load(f)
    content['file_version'] = content['file_version'][:-1] + "000Z"
    with open(descriptor, 'w') as f:
        json.dump(content, f)
```

Further code ran on this because not all file descriptors had decimals

```python
import re
for descriptor in descriptor_filenames:
    with open(descriptor, 'r') as f:
        content = json.load(f)
    if not re.search("[0-9]{4}-[0-9]{2}-[0-9]{2}T[0-9]{2}:[0-9]{2}:[0-9]{2}\.[0-9]{6}Z",content['file_version']):
        split_file_version = content['file_version'].split(":")
        content['file_version'] = f"{split_file_version[0]}:{split_file_version[1]}:{split_file_version[2][:2]}.000{split_file_version[2][2:]}"
        with open(descriptor, 'w') as f:
            json.dump(content, f)
```

Modify file_descriptors file_name to point to 6d file_names (With TRY-CATCH for files without decimals in timestamp)

```python
for descriptor in descriptor_filenames:
    with open(descriptor, 'r') as f:
        content = json.load(f)
    try:
        timestamp_span = re.search("[0-9]{4}-[0-9]{2}-[0-9]{2}T[0-9]{2}:[0-9]{2}:[0-9]{2}\.[0-9]{3}Z", content['file_name']).span()
        content['file_name'] = content['file_name'][:timestamp_span[1] - 1] + "000Z" + content['file_name'][timestamp_span[1]:]
    except:
        timestamp_span = re.search("[0-9]{4}-[0-9]{2}-[0-9]{2}T[0-9]{2}:[0-9]{2}:[0-9]{2}Z", content['file_name']).span()
        content['file_name'] = content['file_name'][:timestamp_span[1] - 1] + ".000000Z" + content['file_name'][timestamp_span[1]:]
    with open(descriptor, 'w') as f:
        json.dump(content, f)
```

## Validation

Validation points:
* All file_descriptors have “file_version” and “file_name”s fields with 6 decimals
* All file_descriptors have “size” and “dcp-type” fields that are non null
* All file_descriptors have all required fields (Includes schema_type)


```python
required_fields_descriptor = {
   "describedBy",
   "schema_type",
   "content_type",
   "size",
   "sha256",
   "crc32c",
   "file_id",
   "file_version",
   "file_name"
}


def correct_timestamp(content):
   timestamps_merged = f"{content['file_name']}_{content['file_version']}"
   if not re.search("^[a-z0-9]{8}(-[a-z0-9]{4}){3}-[a-z0-9]{12}_[0-9]{4}-[0-9]{2}-[0-9]{2}T[0-9]{2}:[0-9]{2}:[0-9]{2}"
                    "\.[0-9]{6}Z_.*\.(gz|pdf)_[0-9]{4}-[0-9]{2}-[0-9]{2}T[0-9]{2}:[0-9]{2}:[0-9]{2}\.[0-9]{6}Z$",
                    timestamps_merged):
       return False
   return True


def size_type_filled(content):
   return False if not content['size'] or not content['content_type'] else True


def all_required_fields(content):
   return False if required_fields_descriptor - set(content.keys()) else True


def main():
   descriptor_filenames = []
   for root, dir, files in os.walk('descriptors'):
       descriptor_filenames.extend([os.path.join(root, file) for file in files])
   for descriptor in descriptor_filenames:
       with open(descriptor, 'r') as f:
           content = json.load(f)
       if not correct_timestamp(content):
           print(descriptor)
       if not size_type_filled(content):
           print(descriptor)
       if not all_required_fields(content):
           print(descriptor)


if __name__ == '__main__':
   main()
```
